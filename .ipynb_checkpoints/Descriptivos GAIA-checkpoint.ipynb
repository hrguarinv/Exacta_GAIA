Histograma classificando por ano

```python
df_H2S = df[df['H2S']<10]
print(df['H2S'].index)
#df_H2S['H2S'][df_H2S['H2S'].index<datalimite]
```

```python
datamin = datetime.datetime(2016, 1, 1)
datamax = datetime.datetime(2017, 1, 1)
array2016= df_H2S['H2S'][df_H2S['H2S'].index<datamax]
datamin = datetime.datetime(2017, 1, 1)
datamax = datetime.datetime(2018, 1, 1)
array2017= df_H2S['H2S'][df_H2S['H2S'].index<datamax]
array2017= array2017[array2017.index>=datamin]
datamin = datetime.datetime(2018, 1, 1)
datamax = datetime.datetime(2019, 1, 1)
array2018= df_H2S['H2S'][df_H2S['H2S'].index<datamax]
array2018= array2018[array2018.index>=datamin]
datamin = datetime.datetime(2019, 1, 1)
datamax = datetime.datetime(2020, 1, 1)
array2019= df_H2S['H2S'][df_H2S['H2S'].index<datamax]
array2019= array2019[array2019.index>=datamin]
```

```python
array2016 = array2016.dropna()
array2017 = array2017.dropna()
array2018 = array2018.dropna()
array2019 = array2019.dropna()
x = [array2016,array2017,array2018,array2019]
fig, axes = plt.subplots(nrows=1, ncols=1)
ax0 = axes
n_bins=5
colors = ['red', 'tan', 'lime','blue']
labels = ['2016','2017','2018','2019']
ax0.hist(x, n_bins, density=True, histtype='bar', color=colors, label=labels)
ax0.legend(prop={'size': 15})
ax0.set_title('Histograma por classe')
```
